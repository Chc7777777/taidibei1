```python
import pandas as pd
```

```python
siji =  pd.read_excel('./3年1，2，3月均值.xlsx')
final_test_df = pd.read_csv('./predict_sku1.csv')
ri = pd.read_csv('./日粒度_14天滑动_三个月预测值.csv')
zhou = pd.read_csv('./周_4周滑动_三个月预测值.csv')
yue = pd.read_csv('./月度_3滑动_未加特征订单需求量.csv')
siji=siji.astype(int)
```

```python
def set_index(data_first):
    data_second = data_first.iloc[:,:]
    data_second=data_second.rename(columns={'order_date':'订单日期','sales_region_code':'销售区域编码','item_code':'产品编码',
                                           'first_cate_code':'产品大类编码','second_cate_code':'产品细类编码',
                                            'sales_chan_name':'销售渠道名称','item_price':'产品价格','ord_qty':'订单需求量'})
#     data_second.columns=['订单日期','销售区域编码','产品编码','产品大类编码','产品细类编码','销售渠道名称','产品价格','订单需求量']
    return data_second
final_test_df=set_index(final_test_df)
siji=pd.merge(final_test_df,siji,on=['销售区域编码','产品编码','产品大类编码','产品细类编码'],how='inner')
siji
```

```python
from pyecharts import options as opts
from pyecharts.charts import Line

# 构造数据

def threeline(titlename,x_data,y_data1,name1,y_data2,name2,y_data3,name3,y_data4,name4):
    x_data = list(x_data)
    y_data1 = list(y_data1)
    y_data2 = list(y_data2)
    y_data3 = list(y_data3)
    y_data4 = list(y_data4)
# 绘制折线图
    line = (
        Line(opts.InitOpts(width="1700px",height="750px"))
        .add_xaxis(x_data)
        .add_yaxis("%s"%name1, y_data1, is_smooth=True, symbol="circle", symbol_size=7, linestyle_opts=opts.LineStyleOpts(width=3))
        .add_yaxis("%s"%name2, y_data2, is_smooth=True, symbol="circle", symbol_size=7, linestyle_opts=opts.LineStyleOpts(width=3))
        .add_yaxis("%s"%name3, y_data3, is_smooth=True, symbol="circle", symbol_size=7, linestyle_opts=opts.LineStyleOpts(width=3))
        .add_yaxis("%s"%name4, y_data4, is_smooth=True, symbol="circle", symbol_size=7, linestyle_opts=opts.LineStyleOpts(width=3))
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(title_opts=opts.TitleOpts(title=titlename), xaxis_opts=opts.AxisOpts(name="月份"), yaxis_opts=opts.AxisOpts(name="订单数"))
    )
    
    # 保存图表到本地文件
    return line
threeline = threeline('月份时间段与订单需求量折线图',ri.index,
                      ri['1月订单需求量'],'日粒度1月初订单需求量',
                      zhou['1月订单需求量'],'周粒度1月需求量',
                      yue['1月订单需求量'],'月粒度1月订单需求量',
                     siji['3年1月均值'],'3年1月均值')
threeline.render('各粒度1月份折线图.html')
threeline.render_notebook()
```
