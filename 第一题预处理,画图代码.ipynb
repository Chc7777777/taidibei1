```python
def concat_groupby(data_,columns_name):
    data_group_by = data_.groupby(columns_name)
    big_code = list(data_group_by.groups)
    big_num_code = list(data_group_by.size())
    new_grouby=pd.DataFrame()
    for i in big_code:
        sort_example=data_group_by.get_group(i).sort_values(by='产品价格')
        new_grouby=pd.concat([new_grouby,sort_example],axis = 0)
    return new_grouby,big_code
```

```python
from pyecharts import options as opts
from pyecharts.charts import *
from pyecharts.globals import *

def bingtu(data1,name):
    data = data1
    
    # 创建 Pie 实例
    pie = Pie(init_opts=opts.InitOpts(theme=ThemeType.INFOGRAPHIC,
                                page_title="Pie- RoseType"))
    
    # 添加数据
    pie.add(series_name="订单需求占比", data_pair=data,radius="50%",  # 饼图
            center=["50%", "50%"],
    
            # 是否展示成南丁格尔图，通过半径区分数据大小
    #         rosetype="radius",  # 通过半径区分数据大小
            rosetype="area",  # 通过面积区分数据大小
    
            is_clockwise=True,  # 扇区排布方向
            label_opts=opts.LabelOpts(is_show=True, position="center"))
    
    # 设置全局配置项
    pie.set_global_opts(
        
        title_opts=opts.TitleOpts(title="{}订单需求占比图".format(name), pos_left="center", title_textstyle_opts=opts.TextStyleOpts(font_size=30)),
        legend_opts=opts.LegendOpts(is_show=True, orient="vertical", pos_left="left", pos_top="middle", item_width=30, item_height=20, textstyle_opts=opts.TextStyleOpts(font_size=20)),
        tooltip_opts=opts.TooltipOpts(trigger="item", formatter="{a} <br/>{b}: {c} ({d}%)"),
    )
    
    # 设置系列配置项
    pie.set_series_opts(
        tooltip_opts=opts.TooltipOpts(
                trigger="item", formatter="{a} <br/>{b}: {d}%"
            ),
        label_opts=opts.LabelOpts(formatter="{b}: {d}%", font_size=14, color="rgba(0, 0, 0, 0.7)"),
    )
    
    # 显示图表
    return pie
```

# 1.2细类区域折线图

```python
# 读取数据集
df = data_second.copy(deep=True)

# 按产品大类编码分组，对订单需求量求和
grouped1 = df.groupby("产品细类编码")["订单需求量"].sum()

# 创建产品大类编码列表
product_categories = list(grouped1.index)

# 按产品大类编码和销售区域编码分组，对订单需求量求和
grouped2 = df.groupby(["产品细类编码", "销售区域编码"])["订单需求量"].sum()

# 将以销售区域编码为键，将每组产品大类中每个销售区域编码的当前订单需求量结合一起，如 {第一个销售区域编码: [第一种产品大类订单需求量，第二种产品大类订单需求量，接下如此]}，作为字典赋给一个新的变量
result = {}
for region in set(df["销售区域编码"]):
    result[region] = []
    for category in product_categories:
        if (category, region) in grouped2.index:
            result[region].append(int(grouped2[(category, region)]))
        else:
            result[region].append(0)

print(result)
```

```python
from pyecharts import options as opts
from pyecharts.charts import *
from pyecharts.globals import *


def anybar(product_categories,result,titlename,x_name,y_name):
    bar_2 = (
        Bar(init_opts=opts.InitOpts(page_title="Bar with MarkPoint", theme=ThemeType.VINTAGE))
        .add_xaxis(product_categories)
        .set_global_opts(
            title_opts=opts.TitleOpts(title=titlename),
            legend_opts=opts.LegendOpts(pos_left="right", pos_top="top", orient="vertical"),
            xaxis_opts=opts.AxisOpts(name=x_name,splitline_opts=opts.SplitLineOpts(is_show=True, linestyle_opts=opts.LineStyleOpts(type_="dashed"))),
            yaxis_opts=opts.AxisOpts(name=y_name,splitline_opts=opts.SplitLineOpts(is_show=True, linestyle_opts=opts.LineStyleOpts(type_="dashed"))),
        )
        .set_series_opts(
            label_opts=opts.LabelOpts(is_show=False),  # 关闭所有值标签的显示
            markpoint_opts=opts.MarkPointOpts(
                data=[
                    opts.MarkPointItem(type_="max", name="最大值"),
                    opts.MarkPointItem(type_="min", name="最小值"),
                ]
            ),
            markline_opts=opts.MarkPointOpts(
                data=[
                    opts.MarkLineItem(type_="average", name="平均值"),
                    opts.MarkLineItem(type_="max", name="最大值"),
                ],
                label_opts=opts.LabelOpts(font_size=12, formatter="{b}{c}s")
            )
        )
    
    )
    
    for key in result:
        bar_2.add_yaxis('%d'%key, result[key],label_opts=opts.LabelOpts(is_show=False))
    
    return bar_2
bar_3=anybar(product_categories,result,'产品细类各地区订单需求量','产品细类编码','订单需求量')
bar_3.render('产品细类各地区订单需求量.html')
bar_3.render_notebook()
```

# 1.5 月份时间段与订单需求量折线图

```python
# 读取数据集到DataFrame
data = data_second.copy(deep=True)

# 将订单日期的数据类型转变为datetime
data['订单日期'] = pd.to_datetime(data['订单日期'])

# 分别按订单日期分类，求和每个时间段的订单需求量
data_start = data[(data['订单日期'].dt.day <= 10)]
grouped_data_start = data_start.groupby([data_start['订单日期'].dt.strftime('%Y-%m'),data_start['销售区域编码'],data_start['产品编码'],data_start['产品大类编码'],data_start['产品细类编码'],data_start['销售渠道名称']])['订单需求量'].sum().reset_index()

data_mid = data[(data['订单日期'].dt.day > 10) & (data['订单日期'].dt.day <= 20)]
grouped_data_mid = data_mid.groupby([data_mid['订单日期'].dt.strftime('%Y-%m'),data_mid['销售区域编码'],data_mid['产品编码'],data_mid['产品大类编码'],data_mid['产品细类编码'],data_mid['销售渠道名称']])['订单需求量'].sum().reset_index()

data_end = data[data['订单日期'].dt.day > 20]
grouped_data_end = data_end.groupby([data_end['订单日期'].dt.strftime('%Y-%m'),data_end['销售区域编码'],data_end['产品编码'],data_end['产品大类编码'],data_end['产品细类编码'],data_end['销售渠道名称']])['订单需求量'].sum().reset_index()

# 按月份分组，对每个数据集中每个月份的订单需求量求和

monthly_start_data = grouped_data_start.groupby('订单日期')['订单需求量'].sum().reset_index()
monthly_mid_data = grouped_data_mid.groupby('订单日期')['订单需求量'].sum().reset_index()
monthly_end_data = grouped_data_end.groupby('订单日期')['订单需求量'].sum().reset_index()

# 生成新的DataFrame，订单日期格式为'YYYY-MM'
new_data = pd.DataFrame({'订单日期': monthly_start_data['订单日期'], '月初订单需求量': monthly_start_data['订单需求量'], '月中订单需求量': monthly_mid_data['订单需求量'], '月末订单需求量': monthly_end_data['订单需求量']})

# 将处理好的数据输出到CSV文件
new_data.fillna(0, inplace=True)
new_data['月初订单需求量'] = new_data['月初订单需求量'].astype(int)
new_data['月中订单需求量'] = new_data['月中订单需求量'].astype(int)
new_data['月末订单需求量'] = new_data['月末订单需求量'].astype(int)
new_data
```

```python
from pyecharts import options as opts
from pyecharts.charts import Line

# 构造数据

def threeline(titlename,x_data,y_data1,name1,y_data2,name2,y_data3,name3):
    x_data = list(x_data)
    y_data1 = list(y_data1)
    y_data2 = list(y_data2)
    y_data3 = list(y_data3)
# 绘制折线图
    line = (
        Line()
        .add_xaxis(x_data)
        .add_yaxis("%s"%name1, y_data1, is_smooth=True, symbol="circle", symbol_size=8, linestyle_opts=opts.LineStyleOpts(width=3))
        .add_yaxis("%s"%name2, y_data2, is_smooth=True, symbol="circle", symbol_size=8, linestyle_opts=opts.LineStyleOpts(width=3))
        .add_yaxis("%s"%name3, y_data3, is_smooth=True, symbol="circle", symbol_size=8, linestyle_opts=opts.LineStyleOpts(width=3))
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(title_opts=opts.TitleOpts(title=titlename), xaxis_opts=opts.AxisOpts(name="月份"), yaxis_opts=opts.AxisOpts(name="订单数"))
    )
    
    # 保存图表到本地文件
    return line
threeline = threeline('月份时间段与订单需求量折线图',new_data['订单日期'],
                      new_data['月初订单需求量'],'月初订单需求量',
                     new_data['月中订单需求量'],'月中订单需求量',
                     new_data['月末订单需求量'],'月末订单需求量')
threeline.render('月份时间段与订单需求量折线图.html')
threeline.render_notebook()
```

```python
workdays = pd.DataFrame(chinese_calendar.get_workdays(start_time,end_time))
workdays = workdays.rename(columns={0:'日期'})
workdays['属性'] = '工作日'
workdays['日期'] = pd.to_datetime(workdays['日期'])
workdays
```

# 1.7 促销期间订单需求量

```python
from pyecharts.charts import Bar
from pyecharts import options as opts

def discount():
    # 读取数据集
    data = data_second.copy(deep=True)
    
    # 将订单日期转换为datetime格式
    data['订单日期'] = pd.to_datetime(data['订单日期'], format='%Y/%m/%d')
    data['促销期'] = ''
    for i in range(np.sort(list(set(data['订单日期'].dt.year)))[0],np.sort(list(set(data['订单日期'].dt.year)))[-1]+1):
        data.loc[(data['订单日期'] >= '%d-03-01'%i) & (data['订单日期'] <= '%d-03-08'%i), '促销期'] = '38女神节'
        data.loc[(data['订单日期'] >= '%d-06-01'%i) & (data['订单日期'] <= '%d-06-20'%i), '促销期'] = '618'
        data.loc[(data['订单日期'] >= '%d-10-01'%i) & (data['订单日期'] <= '%d-10-07'%i), '促销期'] = '国庆节'
        data.loc[(data['订单日期'] >= '%d-11-01'%i) & (data['订单日期'] <= '%d-11-11'%i), '促销期'] = '双十一'
        data.loc[(data['订单日期'] >= '%d-12-01'%i) & (data['订单日期'] <= '%d-12-12'%i), '促销期'] = '双十二'
        data.loc[(data['订单日期'] >= '%d-12-24'%i) & (data['订单日期'] <= '%d-12-25'%i), '促销期'] = '圣诞节'
        data.loc[(data['订单日期'] >= '%d-01-01'%i) & (data['订单日期'] <= '%d-02-10'%i), '促销期'] = '年货节'
    
    def a(x):
        if x=='':
            x='非促销日'
            
        return x 
    data['促销期'] = data['促销期'].apply(a)
    
    # 根据“促销期”和“销售年份”分组，计算每个促销期的订单需求量
    sale_data = data.groupby(['促销期', data['订单日期'].dt.year])['订单需求量'].mean().reset_index()
    
    # 根据“促销期”分组，计算多年的平均值
    yearly_sale_data = sale_data.groupby('促销期')['订单需求量'].mean().round(0).reset_index()
    
    # 输出结果
    
    
    #print(sale_data)
    print(yearly_sale_data)
    
    
    # 绘制柱状图
    bar3 = (
        Bar(init_opts=opts.InitOpts(page_title="Bar with MarkPoint", theme=ThemeType.VINTAGE))
        .add_xaxis(yearly_sale_data['促销期'].tolist())
        .add_yaxis('订单需求量', yearly_sale_data['订单需求量'].tolist())
        .set_global_opts(title_opts=opts.TitleOpts(title='促销期日均订单需求量'))
    )
    
    # 输出结果
    bar3.render('促销期日均订单需求量.html')
    
    return bar3
bar_discount=discount()
bar_discount.render_notebook()
```

# 1.8 季节订单需求量

```python
import pandas as pd
from IPython.display import display
from pyecharts.options import TitleOpts,LabelOpts,LegendOpts,ToolboxOpts,VisualMapOpts
# 读取数据集
def season():
    data = data_second.copy(deep=True)
    
    # 将订单日期转换为datetime格式
    data['订单日期'] = pd.to_datetime(data['订单日期'], format='%Y/%m/%d')
    # 按季节时间段分组，计算每个季节的总订单需求量
    season_sale_data = data.groupby(pd.Grouper(key='订单日期', freq='Q'))['订单需求量'].sum().reset_index()
    
    # 处理数据，将季节表示为“某年春季/某年夏季/某年秋季/某年冬季”的形式
    season_sale_data['季节'] = season_sale_data['订单日期'].dt.strftime('%Y').astype(str) + '年' + \
                              season_sale_data['订单日期'].dt.quarter.apply(lambda x: 
                                    {1: '春季', 2: '夏季', 3: '秋季', 4: '冬季'}.get(x)) 
    
    display(season_sale_data)
    from pyecharts.charts import Line
    from pyecharts import options as opts
    
    # 绘制折线图
    line = (
        Line()
        .add_xaxis(season_sale_data['季节'].tolist())
        .add_yaxis('订单需求量', season_sale_data['订单需求量'].tolist())
        .set_global_opts(title_opts=opts.TitleOpts(title='季节订单需求量折线图', subtitle='按季节时间段分组'),
                        xaxis_opts=opts.AxisOpts(name_rotate=60,axislabel_opts={"rotate":45}))
    )
    
    # 输出结果
    return line
season_line=season()
season_line.render('季节订单需求量折线图.html')
season_line.render_notebook()
```
